### Epoch: 500, Local Epoch: 2, LR: 0.1, Noise Rate: 0.2

```python
# Symmetric Noise
noise_rate = 0.2
f = "fedLNL_cifar_mobile_500_C[0.1]_BS[50]_LE[2]_IID[True]_LR[0.1]_MMT[0.5]_NT[symmetric]_NGN[[100]]_GNR[[0.2]]_PT[label2].csv"
noise_acc = pd.read_csv(result_dir + "No_LNL_ver2/symmetric/" + f)['test_acc']

x = range(1, 501)
plt.plot(x, noise_acc, label='No LNL({})'.format(noise_rate))
print('No LNL({})'.format(noise_rate), round(max(noise_acc), 2), round(noise_acc[490:].mean(), 2))

# Co-Teaching
f = "coteaching_cifar_mobile_500_C[0.1]_BS[50]_LE[2]_IID[True]_LR[0.1]_MMT[0.5]_NT[symmetric]_NGN[[100]]_GNR[[0.2]]_PT[label2]_Tk[100].csv"
noise_acc = pd.read_csv(result_dir + "coteaching/symmetric/" + f)['test_acc1']

x = range(1, 501)
plt.plot(x, noise_acc, label='Co-Teaching({})'.format(noise_rate))
print('Co-Teaching({})'.format(noise_rate), round(max(noise_acc), 2), round(noise_acc[490:].mean(), 2))

# GMix
f = "fedGMixcifar_mobile_500_C[0.1]_BS[50]_LE[2]_IID[True]_LR[0.1]_MMT[0.5]_NT[symmetric]_NGN[[100]]_GNR[[0.2]]_PT[label2].csv"
noise_acc = pd.read_csv(result_dir + "GMix/symmetric/" + f)['test_acc']

x = range(1, 195)
plt.plot(x, noise_acc, label='GMix({})'.format(noise_rate))
print('GMix({})'.format(noise_rate), round(max(noise_acc), 2), round(noise_acc[185:].mean(), 2))

# DivideMix
f = "dividemix_cifar_mobile_500_C[0.1]_BS[50]_LE[2]_IID[True]_LR[0.1]_MMT[0.5]_NT[symmetric]_NGN[[100]]_GNR[[0.2]]_PT[label2]_Tk[100].csv"
noise_acc = pd.read_csv(result_dir + "dividemix/symmetric/" + f)['test_acc1']

x = range(1, 501)
plt.plot(x, noise_acc, label='DivideMix({})'.format(noise_rate))
print('DivideMix({})'.format(noise_rate), round(max(noise_acc), 2), round(noise_acc[490:].mean(), 2))

# Hard Label
f = "HB_H_cifar_mobile_500_C[0.1]_BS[50]_LE[2]_IID[True]_LR[0.1]_MMT[0.5]_NT[symmetric]_NGN[[100]]_GNR[[0.2]]_PT[label2].csv"
noise_acc = pd.read_csv(result_dir + "HS/symmetric/" + f)['test_acc']

x = range(1, 122)
plt.plot(x, noise_acc, label='HH({})'.format(noise_rate))
print('HH({})'.format(noise_rate), round(max(noise_acc), 2), round(noise_acc[112:].mean(), 2))

# History Hard Label
f = "cifar_mobile_500_C[0.1]_BS[50]_LE[2]_IID[True]_LR[0.1]_MMT[0.5]_NT[symmetric]_NGN[[100]]_GNR[[0.2]]_PT[label2].csv"
noise_acc = pd.read_csv(result_dir + "LG_finetuning/symmetric/" + f)['test_acc']

x = range(1, 501)
plt.plot(x, noise_acc, label='LG FT({})'.format(noise_rate))
print('LG FT({})'.format(noise_rate), round(max(noise_acc), 2), round(noise_acc[490:].mean(), 2))



plt.legend()
plt.show()
print("\n" * 2)
plt.close()
```
